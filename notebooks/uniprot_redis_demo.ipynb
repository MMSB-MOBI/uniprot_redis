```python
from uniprot_redis.store import UniprotStore
```

```python
store = UniprotStore()
```

## Load a proteome in Uniprot Store

For the example, we insert 2 uniprot E.coli proteome, downloaded from here as xml : https://www.uniprot.org/proteomes/UP000000625 and 

```python
ecoli_k12 = store.load_uniprot_xml(file='data/k12_proteome.xml.gz')
```

```python
ecoli_sakai = store.load_uniprot_xml(file='data/sakai_proteome.xml.gz')
```

We can then identify these collections of proteins by keywords we want, in order to retrieve them later

```python
store.save_collection('e_coli_k12', ecoli_k12)
store.save_collection('e_coli_sakai', ecoli_sakai)
```

```python
store.list_collection()
```

## Interrogate uniprot store

#### Have access to a single protein

```python
prot = store.get_protein('P00582')
print(prot.id)
print(prot.full_name)
print(prot.sequence)
print(prot.dict().keys()) # All properties of prot model
```

#### Have access to an entire collection through iterator

```python
for p in store.get_protein_collection('e_coli_k12'):
    print(p)
```

## Other

### Have access to all proteins in all collections

```python
for p in store.proteins:
    print(p)
```

### Erase a collection

```python
store.delete_collection('e_coli_k12')
store.list_collection()
```

### Erase database

```python
store.wipe_all()
```
